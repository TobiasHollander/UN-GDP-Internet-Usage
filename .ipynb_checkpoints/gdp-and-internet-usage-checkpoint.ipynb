```python
gdp = gdp_df.drop(columns=['Value Footnotes'])
```

```python
internet = internet_df.drop(columns=['Value Footnotes'])
internet
```

```python
gdp = gdp.rename(columns={'Country or Area':'Country','Value':'GDP_Per_Capita'})
```

```python
internet = internet.rename(columns={'Country or Area':'Country','Value':'Internet_Users_Pct'})
#internet
```

```python
gdp_and_internet_use = pd.merge(gdp, internet, 
                               left_on =['Country','Year'], right_on =['Country','Year'],
                               how = 'outer')
gdp_and_internet_use
```

## Is there anything surprising or unusual in any of these plots?
<ul>
<li><strong>Andorra</strong> has the lowest internet usage for 2004 in the top_5_internet_trend. One reason could be that internet access was available only through the national telephone company and  by 2010 the whole country was to have Fibre-Optic</li> 
<li><strong>Dem. People's Rep. of Korea</strong> has zero internet usage for all three years in bottom_5_internet_trend. One reason could be that the government has absolute authority over and control of the press and information</li>
<li><strong>Cayman Islands</strong> doesn't have gdp for 2004 in top_5_gdp_trend. One reasons could be <em>Hurricane Ivan</em>.</li>
<li><strong>Central African Republic</strong> has a spike in gdp for 2009 and then drops in 2014 in bottom_5_gdp_trend. One reason could be the Central African Republic Civil War which started in 2012.
</li>
</ul>
